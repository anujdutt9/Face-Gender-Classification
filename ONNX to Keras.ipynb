```python
# Import Dependencies
import onnx
from onnx2keras import onnx_to_keras
```

```python
# Load ONNX model
onnx_model = onnx.load('./vgg_face.onnx')
```

```python
# Call the converter (input - is the main model input name, can be different for your model)
k_model = onnx_to_keras(onnx_model, ['input'], change_ordering=True)
```

```python
k_model.summary()
```

```python
k_model.save('./vgg_face_keras_hwc.h5')
```

```python
k_model.save_weights('./vgg_face_keras_weights_hwc.h5')
```
